```python
#today i just want to write a code that could find cube-roots of numbers
#needs to debug the code sooner or later
inp = int(input('insert a value to verify its cube root'))
for guessvalue in range(abs(inp)+1):
    if guessvalue ** 3 >= abs(inp) :
        break
if guessvalue ** 3 != abs(inp):
    print(str(inp),' is not a perfect cube')
if inp < 0:
    inp = -inp
if guessvalue ** 3 ==  abs(inp):
    print('the cube-root of ',str(inp),'is',str(guessvalue))
```

```python
#we are to write a code that prints the no vowels in an input 
inp = input("insert some wordings to know how many vowels they contain")
#z = [a,e,i,o,u]
count = 0
y= 0
for x in range len(inp):
    if inp[x] = "a" or if inp[x] = "e":
        count += 1
    elif inp[x] = "i" or if inp[x] = "o":
        count += 1
    elif inp[x] = "u":
        count += 1
    else :
        y += 1
print ("there are ", str(count),"vowels and",str(y)," consonants in your input.")
```
